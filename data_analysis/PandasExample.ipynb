Tady už jen pro zájemce - funkce, které mi pomohly vytvořit pro vás testovací csv, nemělo by tady být nic nového, vše už znáte - random, seznamy, slovníky, funkce atd.

```python
import random
import pandas
import matplotlib

names = ["Anna","Klára","Miluše","Cecílie","Libuše", "Dobromila", "Vlastimila", "Lucie", "Lenka"]
surnames = ["Nováková","Svobodová","Novotná","Pospíšilová","Němcová", "Marková", "Černá"]
positions = ["junior","medior","senior","tester"]
salaries = [5000*random.randint(5,30) for _ in range(10)]

def generate_random_person(names, surnames, salaries, positions):
    return {"jmeno":random.sample(names,1)[0],
            "prijmeni":random.sample(surnames,1)[0],
            "plat":random.sample(salaries,1)[0],
            "pozice":random.sample(positions,1)[0]}

def generate_people(k):
    return [generate_random_person(names, surnames, salaries, positions) for _ in range(k)]

zeny = pandas.DataFrame(generate_people(100), columns=["jmeno","prijmeni","plat", "pozice"])

zeny.drop_duplicates().to_csv("nahodne_zeny.csv", index = False)
```
